```python
# Chroma = distribution of pitch classes or musical notes in an audio signal

# Chroma_stft = short time Fourier transform of the audio signal (between 0 and 1)
# quantifies the intensity of each of the 12 pitch classes across different time frames of the audio signal.

# Chroma_stft_mean = mean of the chroma_stft


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Get the unique genres
genres = df['label'].unique()

# Create subplots
fig, axes = plt.subplots(1, len(genres), figsize=(6 * len(genres), 6), sharey=True)

# Plot histogram for each genre
for i, genre in enumerate(genres):
    genre_data = df[df['label'] == genre]
    ax = axes[i]
    ax.hist(genre_data['chroma_stft_mean'], bins=30, alpha=0.5)
    ax.set_xlabel('Chroma STFT')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of Chroma STFT - {genre}')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Get the unique genres
genres = df['label'].unique()

# Create subplots
fig, axes = plt.subplots(1, len(genres), figsize=(6 * len(genres), 6), sharey=True)

# Plot histogram for each genre
for i, genre in enumerate(genres):
    genre_data = df[df['label'] == genre]
    ax = axes[i]
    ax.hist(genre_data['tempo'], bins=30, alpha=0.5)
    ax.set_xlabel('Tempo')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of Tempo - {genre}')

plt.tight_layout()
plt.show()
```

```python

# Create a swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x='label', y='spectral_centroid_mean', data=df)
plt.xlabel('Genre')
plt.ylabel('Spectral Centroid Mean')
plt.title('Swarm Plot of Spectral Centroid Mean by Genre')
plt.show()

```

```python

# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='label', y='spectral_centroid_mean', data=df)
plt.xlabel('Genre')
plt.ylabel('Spectral Centroid Mean')
plt.title('Violin Plot of Spectral Centroid Mean by Genre')
plt.show()
```

```python

# Create a KDE plot
plt.figure(figsize=(10, 6))
for genre, genre_data in df.groupby('label'):
    sns.kdeplot(genre_data['spectral_centroid_mean'], label=genre)
plt.xlabel('Spectral Centroid Mean')
plt.ylabel('Density')
plt.title('KDE Plot of Spectral Centroid Mean by Genre')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(75, 25))
sns.heatmap(df.drop(['filename','label'], axis=1).corr(), cmap='coolwarm', annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
