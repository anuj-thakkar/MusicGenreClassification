```python
# Chroma = distribution of pitch classes or musical notes in an audio signal

# Chroma_stft = short time Fourier transform of the audio signal (between 0 and 1)
# quantifies the intensity of each of the 12 pitch classes across different time frames of the audio signal.

# Chroma_stft_mean = mean of the chroma_stft


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Get the unique genres
genres = df['label'].unique()

# Create subplots
fig, axes = plt.subplots(1, len(genres), figsize=(6 * len(genres), 6), sharey=True)

# Plot histogram for each genre
for i, genre in enumerate(genres):
    genre_data = df[df['label'] == genre]
    ax = axes[i]
    ax.hist(genre_data['chroma_stft_mean'], bins=30, alpha=0.5)
    ax.set_xlabel('Chroma STFT')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of Chroma STFT - {genre}')

plt.tight_layout()
plt.show()
```
