```python
import pandas as pd

pd.set_option('display.max_columns', None)
df = pd.read_csv('data/features_30_sec.csv').drop(['filename'], axis=1)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

label_ohe_encoder = OneHotEncoder(handle_unknown='ignore')
label_dummies = label_ohe_encoder.fit_transform(df[['label']])
label_dummies_df = pd.DataFrame(label_dummies.toarray(), columns=label_ohe_encoder.categories_[0])

df.join(label_dummies_df).drop(['label'], axis=1)
df
```
