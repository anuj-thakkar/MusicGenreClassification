```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('data/features_30_sec.csv').drop(['filename', 'length'], axis=1)
df.head()
```

```python
# Scaling the Feature columns
scaler = StandardScaler()
X = scaler.fit_transform(np.array(df.iloc[:, :-1], dtype=float))
y = np.array(df.iloc[:, -1], dtype=str)

print(X.shape)
```

```python
# Label Encoding the Target column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y.shape)
```

```python
# Applying PCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# plot scree plot of Component Number vs. Explained Variance
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)')
plt.title('Explained Variance')
plt.show()


# show full array and not in scientific notation
np.set_printoptions(threshold=np.inf, suppress=True)

print(X)
```

```python
# PCA with 10 components
pca = PCA(n_components=15)
X = pca.fit_transform(X)
print(X.shape)
print(y.shape)

```

```python

# Save new data to csv
df = pd.DataFrame(X)
df['label'] = y
df.to_csv('data/clean/features_30_sec_pca.csv', index=False)
```
